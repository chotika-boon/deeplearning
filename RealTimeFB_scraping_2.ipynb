```python
import pandas as pd
import facebook_scraper
```

```python
post_ids = ['710692170445823']
options = {"comments": True,
           "reactors": True,
           "allow_extra_requests": True,
}
cookies = "/content/cookies.txt" #it is necessary to generate a Facebook cookie file

replies = []

for post in facebook_scraper.get_posts(post_urls=post_ids, cookies=cookies, options=options):
  for p in post['comments_full']:
    replies.append(p)
```

```python
for i in range(len(replies)):
  replies[i]["comment_time"] = str(replies[i]["comment_time"])
```

```python
actualResult = []
your_keys = ['comment_id','commenter_id','comment_text','comment_time']
for i in range(len(replies)):
  actualResult.append({key: replies[i][key] for key in your_keys})
```
